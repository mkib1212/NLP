# Data Preparation / Cleaning

## 1. Remove punctuation

```python
import string
print(string.punctuation)

print("I am by WBS"  == "I am by WBS.") # False


def remove_punctuation(txt):
    txt_nopunct = "".join([c for c in txt if c not in string.punctuation])
    
    """for c in txt:
        if c not in string.punctuation:
            return c """
    
    return txt_nopunct


df["sms_nopunct"] = df["sms"].apply(lambda x: remove_punctuation(x))

df.head()
```

## 2. Tokenize and Lowering

using REGEX

```python
import re # regex library

def tokenize(txt):
    tokens = re.split("\W+",txt) # split the text with Non-Words, + means one or more charachters
    return tokens 


df["sms_tokens"] = df["sms_nopunct"].apply(lambda x: tokenize(x.lower()))

df.head()
```

## 3. Remove Stop Words
remove words with are not needde ---> lower semantic values

i, me, myself, we, our, you, yours, etc. 

```python
stopwords = nltk.corpus.stopwords.words('english')

print(stopwords[0:10])
print(len(stopwords))

def remove_stopwords(txt_tokenized):
    txt_clean = [word for word in txt_tokenized if word not in stopwords]

    """ 
    mylist = []
    for word in txt_tokenizeed:
        if word not in stopwords:
            mylist.append(word)
    """

    return txt_clean

df["sms_nostopw"] = df["sms_tokens"].apply(lambda x: remove_stopwords(x))

df.head()
```

## Stemming: Convert the words to their root word

example: (code, coder, coders, coding) --> code

benefits:
- to save the long time processing
- reduce the count of words (features)
- correlate words with similar meaning (good, nice, etc.)

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer() 

print(ps.stem("coder"))
print(ps.stem("coders"))

print(ps.stem("coding"))
print(ps.stem("code"))


# Not good
print(ps.stem("goose")) # goos
print(ps.stem("geese")) # gees

def stemming(txt):
    txt_stem = [ps.stem(word) for word in txt]
    return txt_stem


df["stemmed"] = df["sms_nostopw"].apply(lambda x: stemming(x))

df.head()
```

## Lemmatizaition (Simiar to Stemming but more accurate)
- more accurate
- too slow : comupation expensive
- reduces the words to the dictionary word

```python
wn = nltk.WordNetLemmatizer()

print(wn.lemmatize("goose")) #  goose
print(wn.lemmatize("geese")) #  goose

def lemmatization(txt):
    txt_lemmas = [wn.lemmatize(word) for word in txt]
    return txt_lemmas


df["lemmatized"] = df["sms_nostopw"].apply(lambda x: lemmatization(x))

df.head()
```

## Standardization

some texts have abbriviations --> convert back to the full text

```python
# Create a dictionary
loopup_dict = {
    "nlp": "Natural Language Processing",
    "ur" : "your",
    "wbu" : "What about you"
}


def text_abbr(input_text):
    words = input_text.split()
    new_words = []

    for word in words:
        word = re.sub(r"[^\w\s]", "", word)
        if word.lower() in loopup_dict:
            word = loopup_dict[word.lower()]
            new_words.append(word)
            new_text = " ".join(new_words)
    return new_text, new_words 

text = "i like nlp. The is ur choice, wbu ?"

txt_str, new_word_list = text_abbr(text)

print(txt_str)
print(new_word_list)
```

# Count Vectorization

counts the word frequency 

```python
from sklearn.feature_extraction.text import CountVectorizer

# Creat the cv
cv = CountVectorizer()


# 3x rows --> 3x documents
corpus = ["This is a sentence is",
          "This is another sentence",
          "Third document is there"]

# Learn the dictionary from corpus
x = cv.fit(corpus)

# Get the list of words (dictionary)
print(cv.get_feature_names_out()) # ['another' 'document' 'is' 'sentence' 'there' 'third' 'this']

# Create a document Term Matrix(DTM)
x = cv.transform(corpus)

# Create a DataFrame from DTM
df2 = pd.DataFrame(x.toarray(), columns = cv.get_feature_names_out())
df2.head()
```

# TF-IDF (Term Frequency , Inverse Dcoument Frequency)



```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Creat the cv
tfid_vect = TfidfVectorizer()


# 3x rows --> 3x documents
corpus = ["This is a sentence is",
          "This is another sentence",
          "Third document is there"]

# Learn the dictionary from corpus
x = tfid_vect.fit(corpus)

# Get the list of words (dictionary)
print(tfid_vect.get_feature_names_out()) # ['another' 'document' 'is' 'sentence' 'there' 'third' 'this']

# Create a document Term Matrix(DTM)
x = tfid_vect.transform(corpus)

# Create a DataFrame from DTM
df3 = pd.DataFrame(x.toarray(), columns = tfid_vect.get_feature_names_out())
df3.head()
```
